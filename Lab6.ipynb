```python
#a

#creating a List
a = [1,2,3,4,5,5]
print(a)

a = list([1,2,3,4,5,6])
print(a)

#traversing through a list
for x in a:
    print(x)

#Slicing a list
print(a[1:4])
```

```python
#Methods in list

list1 = ["Myil", 1, 2 , 3, 4, 15, 6, 7]

#append() - adds an element at the end of the list
list1.append("Vaughanan")
print(list1)

#insert() - inserts value at a particular position in a list
#By default the position is the end of the list just like append()
list1.insert(3,'Data Science')
print(list1)


#pop() - removes the element at a specified position
#By default the last element stored in the list is removed
list1.pop()
print(list1)
list1.pop(2)
print(list1)


#remove() - removes the first occurrence of the of specified value
list1.remove(15)
print(list1)


#index() - finds the position at the first occurrence of specified value
list1.index('Data Science')


#reverse() - reverses the contents of the list
print(list1)
print("Before reversal: ",id(list1))
list1.reverse()
print(list1)
print("After reversal: ",id(list1))


#clear() - Used to clear a list in Python
lst = [1, 6, 11, 16]
print(lst)
lst.clear()
print(lst)


#Aliasing - It means to refer the same instance with different names. An alias is a a second name for a piece of data.
list3 = list1
print("List 2: ",id(list3))
print("List 1: ",id(list1))


#Cloning - If we want to modify a list and also keep the copy of the original list
list4 = list1[:] #make clones by using slice operator
print("Comparing Values: ",list4 == list1)
print("Comparing Address: ",list4 is list1)


#copy() - if we don't want to create a alias but we need a duplicate value then we can use copy()
list3 = list1.copy()
print("List 2: ",id(list3))
print("List 1: ",id(list1))

#extend() - it adds the specified list elements (or any iterable) to the end of the current list.
list2 = ["Data", "Science"]
list1.extend(list2)
print(list1)
```

```python
#del() - deletes the existence of the variable or deletes the particular value in a list
x = [1,2,3,4,5]
del(x)
```

```python
#b

#creating a Tuple
a = (10,20,30,"orange","mango")
print(a)
type(a)

#accessing a particular element in a tuple
print(a[0]) #printint the value of tuple at the 0th index
print(a[1:3]) #printing values of tuple at the indices 1 and 2
print(a[-1]) #printing only the last element
print(a[:-1]) #last element excluded
print(a[::-1]) #reverse order of a tuple

#Traversing
for x in a:
    print(x)

#Concatenation:
a = (3,6,9)
b = (4,7,9)
c = a+b
print(c)

#Repetition
a = (10,20,30)
b = 3*a
print(b)

#Membership Operations
print(10 in a)
print(80 not in a)

#count(): Checks the total number of occurences of a particular value in a Tuple.
a = (1,2,3,4,5,6,7,7,7,7,7,7,7)
print(a.count(7))

#index(): checks the index of the first occurence of a particular value in a tuple
print(a.index(7))

#min(): It returns the smallest value in a tuple
print(min(a))

#max(): It returns the largest value in a tuple
print(max(a))

#len(): It finds the length of a tuple
print(len(a))

#del(): It is a function used to remove the tuples from the memory
c = (1,2,3,4,5)
del(c)
print("c variable is deleted")
```

```python
#c

#Creating a dictionary
students={ 1:"Avantika", 2:'Chethan', 3:"Harshik", 4:"Hruthuik", 5:'Naveen'}
print(students)

#Accessing a dictionary
a = students[5]
print(a)

#keys(): This method returns the keys of the dictionary, as a list.  
x = students.keys()
print(x)

#values(): This method returns the values of the dictionary, as a list.  
x = students.values()
print(x)

#items(): This method returns the key-value pairs of the dictionary, as tuples in a list.
x = students.items()
print(x)

#update(): This method inserts the specified items to the dictionary.
new_student = {6:"Mithun",7:"Myil Vaughanan V L"}
print(new_student) 
students.update(new_student)
print(students)

#copy(): This method returns a copy of the specified dictionary.
ok = students.copy()
print(ok)

#sorted(): method returns a sorted sequence of the keys in the dictionary as a list
dict1={3:"Three",4:"Four",1:"One",2:"Two"}
dict2=sorted(dict1)
print(dict2)

#pop(): This method is used to remove and display an item from the dictionary. 
x = students.pop(4)
print(x)

#popitem(): This method removes the last item in the dictionary
dict2={3:"Three",4:"Four",1:"One",2:"Two", 5:"Five"}
x=dict2.popitem()
print(x)

#clear(): This method method removes all the elements from a dictionary.
students.clear()
print(students)


```
